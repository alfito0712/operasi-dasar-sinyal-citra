# Operasi Dasar pada Sinyal dan Citra
Notebook untuk tugas Pengolahan Sinyal Digital.

```
import numpy as np
import matplotlib.pyplot as plt
import cv2

plt.rcParams['figure.figsize']=(8,4)
```

## A.1 Membuat Sinyal

```
n=np.arange(0,31)
x1=np.sin(0.5*np.pi*n)
x2=np.where(n<5,0,1)

plt.stem(n,x1)
plt.title('x1[n]')
plt.show()

plt.stem(n,x2)
plt.title('x2[n]')
plt.show()

print('x1: sinyal sinus periodik')
print('x2: sinyal step')
```

## A.2 Penjumlahan

```
y=x1+x2
fig,ax=plt.subplots(3,1,figsize=(8,8))
ax[0].stem(n,x1)
ax[0].set_title('x1')
ax[1].stem(n,x2)
ax[1].set_title('x2')
ax[2].stem(n,y)
ax[2].set_title('y=x1+x2')
plt.tight_layout()
plt.show()

print('Analisis: amplitudo meningkat setelah n>=5.')
```

## A.3 Penggeseran

```
for k in [-3,0,4]:
    plt.figure()
    plt.stem(n,np.sin(0.5*np.pi*(n-k)))
    plt.title(f'Shift k={k}')
    plt.show()
```

## A.4 Amplifikasi

```
for a in [0.5,1,2,-1]:
    plt.figure()
    plt.stem(n,a*x1)
    plt.title(f'alpha={a}')
    plt.show()
```

## B.1 Membaca Citra

```
img=cv2.imread('/mnt/data/WhatsApp Image 2026-04-05 at 14.00.21.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print("Ukuran:",img.shape)
print("Tipe:",img.dtype)
print("Min:",img.min(),"Max:",img.max())
plt.imshow(img)
plt.axis('off')
plt.show()
```

## B.2 Penjumlahan Citra

```
img2=np.flip(img,axis=1)
add=cv2.add(img,img2)

fig,ax=plt.subplots(1,3,figsize=(12,4))
for a,i,t in zip(ax,[img,img2,add],['Asli','Mirror','Penjumlahan']):
    a.imshow(i);a.set_title(t);a.axis('off')
plt.show()
```

## B.3 Penggeseran Citra

```
rows,cols=img.shape[:2]
for dx,dy in [(30,0),(0,30),(30,30)]:
    M=np.float32([[1,0,dx],[0,1,dy]])
    sh=cv2.warpAffine(img,M,(cols,rows))
    plt.figure(figsize=(5,5))
    plt.imshow(sh)
    plt.title(f'Shift ({dx},{dy})')
    plt.axis('off')
    plt.show()
```

## B.4 Amplifikasi Citra

```
for a in [0.5,1,1.5,2]:
    out=np.clip(img.astype(np.float32)*a,0,255).astype(np.uint8)
    plt.figure(figsize=(5,5))
    plt.imshow(out)
    plt.title(f'alpha={a}')
    plt.axis('off')
    plt.show()

    plt.figure()
    plt.hist(out.ravel(),256,[0,256])
    plt.title(f'Histogram alpha={a}')
    plt.show()
```

## C. Uji Sistem Linier

```
def T1(x): return 2*x
def T2(x): return x**2

print('Homogenitas T1')
for a in [0.5,2,-1]:
    print(a,np.allclose(T1(a*x1),a*T1(x1)))

print('Additivitas T1:',np.allclose(T1(x1+x2),T1(x1)+T1(x2)))

print('Homogenitas T2')
for a in [0.5,2,-1]:
    print(a,np.allclose(T2(a*x1),a*T2(x1)))

print('Additivitas T2:',np.allclose(T2(x1+x2),T2(x1)+T2(x2)))
```

## D. Analisis HOTS

- Penjumlahan dan amplifikasi merupakan dasar prinsip superposisi.
- T(x)=2x memenuhi homogenitas dan additivitas sehingga linier.
- T(x)=x² tidak memenuhi kedua sifat tersebut sehingga nonlinier.
- Penggeseran digunakan untuk simulasi delay dan augmentasi citra.
